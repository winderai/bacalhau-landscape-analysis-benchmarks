```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
```

```python
def load_csv(file_path):
    df = (
        pd.read_csv(file_path, skiprows=3)
        .iloc[1: , :]
    )
    df.columns = df.columns.map(lambda x: x.replace("eu-central-1:CWAgent ", "").replace(" Average 60", ""))
    print("NaN check: ", df.isna().sum().sum())
    df = df.set_index('Full label')
    df = df.dropna(axis=0, how='all').dropna(axis=1, how='all')
    print("NaN check: ", df.isna().sum().sum())
    print("shape: ", df.shape)

    return df
```

## cluster-size-1

```python
df_cs1_cpu = load_csv("cs-1-cpu_usage_idle__cpu_usage_iowait__cpu_usage_system-2022_05_10_15_23_00-2022_05_10_16_58_00-UTC.csv")
print("host name / column check: ", any('hadoop' in string for string in df_cs1_cpu.columns.tolist()))
df_cs1_cpu.head()
```

```python
df_cs1_diskio = load_csv("cs-1-diskio_reads__diskio_writes-2022_05_10_15_23_00-2022_05_10_17_16_00-UTC.csv")
print("host name / column check: ", any('hadoop' in string for string in df_cs1_diskio.columns.tolist()))
df_cs1_diskio.head()
```
