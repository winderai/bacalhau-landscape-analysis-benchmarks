# Pandas Word Count

```python
import pandas as pd
```

```python
# load dataset into DataFrame
file_path = "./data/wordcount.txt"

# each entry contains a text row
df = pd.read_csv(file_path, header=None)
df.head()
```

```python
# count words
wc_df = (
    df[0]
    .str
    .split(expand=True) # splits row around ' ' separator and expand into columns
    .stack() # stack columns vertically -> each entry contains a single word
    .value_counts()
    .reset_index()
)
wc_df.head()
```

```python
wc_df.columns = ['Word', 'Frequency'] 
wc_df = wc_df.sort_values(['Frequency', 'Word'], ascending=False) # expensive op - just for 
wc_df
```
